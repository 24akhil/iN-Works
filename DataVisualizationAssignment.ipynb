```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.plot();
```

```python
# dftarget = df[['sepal_length','petal_length']]
```

```python
# Create the figure
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('sepal length', fontsize = 15)
ax.set_ylabel('petal length', fontsize = 15)
ax.set_zlabel('sepal width', fontsize = 15)
ax.set_title('3D Graph', fontsize = 20)

targets = ['sepal_length', 'petal_length', 'sepal_width']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
#     indicesToKeep = iris['target'] == target
    ax.scatter(iris['sepal_length']
               , iris['petal_length']
               , iris['sepal_width']
               , c = color
               , s = 50)
    
ax.legend(targets)
ax.grid()
plt.show()
```
